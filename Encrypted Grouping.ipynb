```python
# Basic importing
# All of our dependencies are listed in a requirements.txt file
from IPython.core.display import display, HTML
import pandas as pd
import numpy as np
import json


#use this version for what was installed via pip
import ghostPii as gp
```

```python
headers = {'Authorization': 'Token 3732725eeb3f1345974479d2479d0202a17ecf88'}
myContext = gp.CryptoContext(headers)
```

```python
with open('demo_data/livedata.json','r') as file:
    locationData = json.load(file)
```

```python
listDF = []
for row in locationData:
    listDF.append(row)
plaintextFrame = pd.DataFrame(listDF)
plaintextFrame = plaintextFrame.astype({'timestamp':str,'latitude':np.float64,'longitude':np.float64})
plaintextFrame
```

```python
encryptedFrame = gp.NormCipherFrame(myContext,plaintextFrame,keyRange=200)
```

```python
newLists = encryptedFrame.group_by(1)
for newList in newLists:
    print(set(newList[1].decrypt()))
    print(newList.decrypt())
```
