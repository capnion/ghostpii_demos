Here we are defining a distance function that is essentially just a euclidean distance function that does a rough conversion to miles for the output

```python
def createDistance(lat,lng,apiContext):
    if len(lat) != len(lng):
        return "cannot use lists of different lengths"
    
    
    # assume euclidean distance formula
    polyString = '(x-a)**2 + (y-b)**2'
    variables = ['x','y','a','b']
    myIndices = []
    myCiphers = []
    
    for i in range(len(lat)):
        myIndices.append((lat.indicesList[i],lng.indicesList[i],lat.indicesList[50],lng.indicesList[50],))
        myCiphers.append((lat.cipherList[i],lng.cipherList[i],lat.cipherList[50],lng.cipherList[50],))
                
    ans = gp.full_polynomial_compute(apiContext,polyString,variables,myIndices,myCiphers,lat.floatData,paillier=True,outPlain=False)
    ans = ans.decrypt()
    
    for k in range(len(ans)):
        # change back to degrees
        ans[k] = (ans[k] ** .5)/10000
        # convert degrees to miles (roughly)
        ans[k] *= 45
    
    return ans
```

```python
def createDistance2(lat,lng,apiContext):
    if len(lat) != len(lng):
        return "cannot use lists of different lengths"
    
    # assume euclidean distance formula
    polyString = '(x-a)**2 + (y-b)**2'
    variables = ['x','y','a','b']
    myIndices = []
    myCiphers = []
    
    for i in range(len(lat)):
        myIndices.append((lat.indicesList[i],lng.indicesList[i],lat.indicesList[50],lng.indicesList[50],))
        myCiphers.append((lat.cipherList[i],lng.cipherList[i],lat.cipherList[50],lng.cipherList[50],))
                
    ans = gp.full_polynomial_compute(apiContext,polyString,variables,myIndices,myCiphers,lat.floatData,paillier=True,outPlain=True)
    #ans = ans.decrypt()
    
    
    for k in range(len(ans)):
        # change back to degrees
        ans[k] = ans[k]**0.5
        # convert degrees to miles (roughly)
        ans[k]*=45
    
    
    return ans
```

```python
def plaintextDistance(lat,lng):
    ans = []
    for i in range(len(lat)):
        num = (lat[i]-lat[50])**2+(lng[i]-lng[50])**2
        ans.append((num**0.5)*45)
    return ans
```

Now we can finally run our distance formula on the encrypted data and output the distance between all cities in miles

```python
leadDistance = createDistance(cipherLat,cipherLng,myContext)
floatDistance = createDistance2(cipherLatFloat,cipherLngFloat,myContext)
groundTruth = plaintextDistance(test_data_lat_float,test_data_lng_float)
```

```python
errors_float = []
errors_int = []
for i in range(len(floatDistance)-1):
    errors_float.append(abs(floatDistance[i]-groundTruth[i])*100/groundTruth[i])
    errors_int.append(abs(leadDistance[i]-groundTruth[i])*100/groundTruth[i])
    #print("Int{}:{}".format(i,abs(leadDistance[i]-groundTruth[i])))
    #print("Float{}:{}".format(i,errors_float[i]))
print(max(errors_int))
print(max(errors_float))
```
